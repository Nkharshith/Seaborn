```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

**EDA Learning Series**

**Part 1 – Univariate Analysis**

**1. Histogram**

```python
sns.histplot(tips["total_bill"], bins=20, kde=True)

plt.title("Histogram")
plt.show()
```

**2. KDE Plot**

```python
sns.kdeplot(tips["total_bill"], fill=True)

plt.title("KDE Plot")
plt.show()
```

**3. Rug Plot**

```python
sns.rugplot(x=tips['total_bill'])
plt.title('Rug plot')

plt.show()
```

**4. ECDF Plot**

```python
sns.ecdfplot(data = tips,x='total_bill')

plt.title('ECDF PLOT')

plt.show()
```

**5. Box Plot**

```python
sns.boxplot(x = tips['total_bill'])
plt.title('Box plot')
plt.show()
```

**6. Violin Plot**

```python
sns.violinplot(x=tips['total_bill'])
```

**7. Count Plot**

```python
sns.countplot(data = tips,x='day')
plt.title('Count plot')
plt.show()
```

**Part 2 – Bivariate Analysis**

**8. Scatter Plot**

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

**9. Line Plot**

```python
import pandas as pd

day_bill = tips.groupby('day')['total_bill'].mean().reset_index()

# plt.plot(day_bill['day'],day_bill['total_bill'],marker='o')

sns.lineplot(data=day_bill,x='day',y='total_bill',marker='o')
plt.show()
```

**10. Area Plot**

```python
day_bill = tips.groupby("day")["total_bill"].sum()

day_bill.plot(kind='area',alpha=0.5)
plt.title("Area Plot")
plt.show()
```

**11. Bar Plot**

```python
tips = sns.load_dataset('tips')
```

```python
sns.barplot(data = tips,x = 'day',y='total_bill',hue='day')
plt.show()
```

**12. Strip Plot**

```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    jitter=0.2,
    hue='day'
)

plt.title("Strip Plot")
plt.show()
```

**13. Swarm Plot**

```python
sns.swarmplot(
    data=tips,
    x="day",
    y="total_bill",
    hue='day'
)

plt.title("Swarm Plot")
plt.show()
```

**14. Box Plot with Categories**

```python
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue = 'day'
)

plt.show()
```

**15. Violin Plot with Categories**

```python
sns.violinplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.show()
```

**16. Regression Plot**

```python
sns.regplot(data=tips,x='total_bill',y='tip')
```

**17. Joint Plot**

```python
sns.jointplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="scatter"
)

plt.show()
```

**18. Hexbin Plot**

```python
plt.hexbin(
    tips["total_bill"],
    tips["tip"],
    gridsize=15,
    cmap="viridis"
)

plt.colorbar(label="Count")
plt.show()
```

**19. 2D Histogram**

```python
plt.hist2d(
    tips["total_bill"],
    tips["tip"],
    bins=20,
    cmap="Blues"
)

plt.colorbar()
plt.show()
```

**Part 3 – Multivariate Analysis**

**20. Heatmap**

```python
corr = tips.corr(numeric_only = True)
sns.heatmap(corr,annot=True,cmap='coolwarm')

plt.show()
```

**21. Pair Plot**

```python
sns.pairplot(tips,hue='sex')
```

**22. FacetGrid**

```python
g = sns.FacetGrid(
    tips,
    col="time"
)

g.map(
    sns.scatterplot,
    "total_bill",
    "tip"
)
```

**23. PairGrid**

```python
g = sns.PairGrid(tips)

g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
```

**24. CATPLOT**

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    kind="box"
)
```

**25. RelPlot**

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex"
)
```

**26. DisPlot**

```python
sns.displot(
    tips,
    x="total_bill",
    kde=True
)
```
